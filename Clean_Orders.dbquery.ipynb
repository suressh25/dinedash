```sql
SELECT
    (SELECT COUNT(DISTINCT order_id) FROM dinedash_streaming.orders_bronze) AS bronze_count,
    (SELECT COUNT(DISTINCT order_id) FROM dinedash_streaming.cleaned_orders) AS cleaned_count;

SELECT
    COUNT(*) AS exploded_rows,
    COUNT(DISTINCT order_id) AS distinct_orders
FROM dinedash_streaming.cleaned_orders;

SELECT
    COUNT(*) AS bronze_rows,
    COUNT(DISTINCT order_id) AS distinct_orders
FROM dinedash_streaming.orders_bronze;

SELECT
    SUM(size(items_ordered)) AS expected_item_rows
FROM dinedash_streaming.orders_bronze;

SELECT
    COUNT(*) AS actual_item_rows
FROM dinedash_streaming.cleaned_orders;
```
