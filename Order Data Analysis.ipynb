```sql
SELECT order_id,customer_id,restaurant_id,ROUND(total_amount,2) AS total_amount
FROM orders_bronze
ORDER BY total_amount DESC
LIMIT 10;
```

```sql
SELECT payment_method,ROUND(AVG(tip),2) as average_tip
FROM orders_bronze
GROUP BY payment_method;
```

```sql
SELECT
    ROUND(
        100 * SUM(
            CASE WHEN tip > total_amount * 0.10 THEN 1
                ELSE 0 
            END
        ) / COUNT(*),2
    ) AS percentage_orders
FROM orders_bronze
```

```sql
SELECT 
    order_status,
    ROUND(100 * COUNT(*) / SUM(COUNT(*)) OVER(),2)
    AS percentage
FROM orders_bronze
GROUP BY order_status
ORDER BY percentage DESC;
```

```sql
SELECT customer_id,count(*) as total_orders
FROM orders_bronze
GROUP BY customer_id
ORDER BY total_orders DESC
LIMIT 10;
```

```sql
SELECT restaurant_id,count(*) as total_orders
FROM orders_bronze
GROUP BY restaurant_id
ORDER BY total_orders DESC
LIMIT 10;
```

```sql
SELECT
    item.item_id,
    item.item_name,
    COUNT(*) AS total_orders
FROM orders_bronze
LATERAL VIEW EXPLODE(items_ordered) exploded AS item
GROUP BY item.item_id,item.item_name
ORDER BY total_orders DESC
LIMIT 10;
```

```sql
SELECT ROUND(AVG(size(items_ordered)),2) AS average_items
FROM orders_bronze;
```

```sql
SELECT restaurant_id,
    item.item_id,
    item.item_name,
    ROUND(SUM(item.price*item.quantity),2) as item_revenue
FROM orders_bronze
LATERAL VIEW EXPLODE(items_ordered) t as item
GROUP BY restaurant_id,item_id,item_name
ORDER BY restaurant_id,item_revenue DESC;
```
