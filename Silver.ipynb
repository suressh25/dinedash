```sql
CREATE OR REPLACE TABLE customers_filtered AS
SELECT *
FROM customers_bronze
WHERE customer_id IS NOT NULL 
    AND location_id IS NOT NULL
    AND signup_date IS NOT NULL
    AND dob IS NOT NULL
    AND email IS NOT NULL 
    AND name IS NOT NULL
    AND _rescued_data IS NULL;
```

```sql
SELECT * FROM customers_filtered
```

```sql
SELECT 
    (SELECT COUNT(*) FROM customers_bronze) AS bronze_count,
    (SELECT COUNT(*) FROM customers_filtered) AS filtered_count,
    (SELECT COUNT(*) FROM customers_bronze) -
    (SELECT COUNT(*) FROM customers_filtered) AS records_removed;
```

```sql
SELECT *
FROM customers_bronze
WHERE customer_id IS NULL 
    OR location_id IS NULL
    OR signup_date IS NULL
    OR dob IS NULL
    OR email IS NULL 
    OR name IS NULL
    OR _rescued_data IS NOT NULL;
```
